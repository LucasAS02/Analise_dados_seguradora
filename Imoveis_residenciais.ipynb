Solicitação para criação de um novo dataframe contendo somente imóveis residenciais.

## IMPORTS

```python
import pandas as pd
```

```python
dataset = pd.read_csv('dados_imoveis.csv', sep = ';')
```

## DATAFRAME

```python
dataset
```

## RELATÓRIO II - IMÓVEIS RESIDENCIAIS

```python
list(dataset['Tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
```

```python
selecao = dataset['Tipo'].isin(residencial)
```

```python
dados_residencial = dataset[selecao]
```

```python
dados_residencial
```

```python
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
dados_residencial.index = range(dados_residencial.shape[0])
```

```python
dados_residencial
```

## EXPORTANDO A NOVA BASE DE DADOS

```python
dados_residencial.to_csv('dados_imoveis_res.csv', sep = ';', index = False)
```
