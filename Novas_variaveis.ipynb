
Solicitação para a criação de 04 novas variáveis: valor bruto, valor do m², valor bruto do m² e uma variável para agregar os tipos de imóveis.

## IMPORTS

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados_resid_tratado.csv', sep = ';')
```

## DATAFRAME

```python
dados
```

## RELATÓRIO V - CRIANDO NOVAS VARIÁVEIS

Valor Bruto

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

Valor do m²

```python
dados['Valor m²'] = dados['Valor'] / dados['Area']
```

```python
dados['Valor m²'] = dados['Valor m²'].round(2)
```

```python
dados.head(10)
```

Valor bruto do m²

```python
dados['Valor Bruto m²'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.head(10)
```

Criando uma nova variável para agregar casa e apartamentos.

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```
